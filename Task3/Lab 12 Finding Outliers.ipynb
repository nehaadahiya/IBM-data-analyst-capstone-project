<h3>Step 3: Identify High Compensation Outliers</h3>


Identify respondents with extremely high yearly compensation.

- Calculate basic statistics (mean, median, and standard deviation) for `ConvertedCompYearly`.

- Identify compensation values exceeding a defined threshold (e.g., 3 standard deviations above the mean).


```python
##Write your code here
mean_comp = df['ConvertedCompYearly'].mean()
median_comp = df['ConvertedCompYearly'].median()
std_comp = df['ConvertedCompYearly'].std()

print("Basic Statistics for Yearly Compensation:")
print(f"Mean: {mean_comp:,.2f}")
print(f"Median: {median_comp:,.2f}")
print(f"Standard Deviation: {std_comp:,.2f}")

#Threshold: 3 standard deviations above the mean
threshold = mean_comp + (3*std_comp)

outliers = df[df['ConvertedCompYearly'] > threshold]

print(f"\nNumber of respondents with extremely high compensation: {len(outliers)}")
print(f"Threshold for high compensation: {threshold:,.2f}")

# Display top 5 extreme cases
print("\nTop 5 Highest Compensations:")
print(outliers[['ConvertedCompYearly', 'Employment', 'Country', 'YearsCodePro']].sort_values(by='ConvertedCompYearly', ascending=False).head())
```

<h3>Step 4: Detect Outliers in Compensation</h3>


Identify outliers in the `ConvertedCompYearly` column using the IQR method.

- Calculate the Interquartile Range (IQR).

- Determine the upper and lower bounds for outliers.

- Count and visualize outliers using a box plot.


```python
##Write your code here
comp_data = df['ConvertedCompYearly'].dropna()

Q1 = comp_data.quantile(0.25)
Q3 = comp_data.quantile(0.75)
IQR = Q3-Q1

#upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#outliers
outliers = comp_data[(comp_data < lower_bound) | (comp_data > upper_bound)]

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Number of Outliers: {len(outliers)}")

#box plot
plt.figure(figsize=(10,6))
sns.boxplot(x=comp_data)
plt.title("Boxplot of Yearly Compensation (ConvertedCompYearly) with Outliers")
plt.xlabel("Yearly Compensation")
plt.show()
```

<h3>Step 5: Remove Outliers and Create a New DataFrame</h3>


Remove outliers from the dataset.

- Create a new DataFrame excluding rows with outliers in `ConvertedCompYearly`.
- Validate the size of the new DataFrame.


```python
##Write your code here
df_no_outliers = df[(df['ConvertedCompYearly']>=lower_bound) & (df['ConvertedCompYearly']<=upper_bound)]

#valiadting the size of new data frame

print(f"Original dataset size: {df.shape[0]} rows")
print(f"Cleaned Dataset size: {df_no_outliers.shape[0]} rows")
print(f"Rows Removed: {df.shape[0] - df_no_outliers.shape[0]}")

```
