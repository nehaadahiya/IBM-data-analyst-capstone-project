### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here

#shape of data 
print("Shape of dataset:", df.shape)

#column names 
print("\nColumn names:")
print(df.columns.tolist())

#data types 
print("\nData types and Non-Null counts in the dataset:")
print(df.info())

#summary- numeric columns
print("\nSummary of the dataset:")
print(df.describe())

#summary- categorical columns
print("\nCategorical Column Summary:")
print(df.describe(include=object))
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here

#identifying missing values
print("Missing values per column:\n", df.isnull().sum())
#percentage of missing values
print("Percentage of missing values per column:\n", (df.isnull().sum() / len(df)) * 100)

#Resolving missing values 
#moderate missing values <18%
df['RemoteWork'].fillna(df['RemoteWork'].mode()[0], inplace=True)
df['SurveyLength'].fillna(df['SurveyLength'].mode()[0], inplace=True)
df['SurveyEase'].fillna(df['SurveyEase'].mode()[0], inplace=True)
#high missing values 
# Median for skewed numeric columns
df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].median(), inplace=True)

# Mean for normally distributed numeric column
df['JobSatPoints_11'].fillna(df['JobSatPoints_11'].mean(), inplace=True)

# Mode for categorical column
df['JobSat'].fillna(df['JobSat'].mode()[0], inplace=True)
```

```python
print("Missing values per column:\n", df.isnull().sum())
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
key_cols = ['Employment','JobSat','YearsCodePro']
for col in key_cols:
    print("f{Distribution for {col}:")
    print(df[col].value_counts(dropna=False))
    print("\n" + "="*50 + "\n")
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns
#KDE plot
plt.figure(figsize=(8,6))
sns.kdeplot(data=df, x='JobSat', fill =True, color='skyblue')
plt.title("Job satisfaction (KDE Plot)")
plt.xlabel("Job Satisfaction ")
plt.ylabel("Density")
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
# Split the comma-separated languages into lists
have = df['LanguageHaveWorkedWith'].dropna().str.split(';').explode()
want = df['LanguageWantToWorkWith'].dropna().str.split(';').explode()

# Count frequencies
have_counts = have.value_counts()
want_counts = want.value_counts()

# Combine into one DataFrame
lang_compare = pd.DataFrame({
    'Have Worked With': have_counts,
    'Want To Work With': want_counts
}).fillna(0).astype(int)

# Plot grouped bar chart
lang_compare.sort_values(by="Have Worked With", ascending=False).head(15).plot(
    kind='bar',
    figsize=(12,6),
    width=0.8
)

plt.title("Comparison of Programming Languages: Have Worked With vs Want To Work With")
plt.ylabel("Number of Respondents")
plt.xlabel("Programming Languages")
plt.xticks(rotation=45, ha="right")
plt.legend(title="")
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Count responses per country
country_counts = df["Country"].value_counts().head(15).index

# Filter dataset for top 10 countries
df_top = df[df["Country"].isin(country_counts)]

# Group data by Country and RemoteWork
remote_region_counts = df_top.groupby(["Country", "RemoteWork"]).size().reset_index(name="Count")

# # --- Grouped Bar Chart ---
# plt.figure(figsize=(12,6))
# sns.barplot(data=remote_region_counts, x="Country", y="Count", hue="RemoteWork")
# plt.title("Distribution of Remote Work by Top 15 Countries", fontsize=14)
# plt.xticks(rotation=45, ha="right")
# plt.ylabel("Number of Respondents")
# plt.xlabel("Country")
# plt.legend(title="Remote Work Type")
# plt.tight_layout()
# plt.show()

# --- Heatmap ---
pivot_table = remote_region_counts.pivot(index="Country", columns="RemoteWork", values="Count").fillna(0)

plt.figure(figsize=(13,8))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="Blues")
plt.title("Heatmap of Remote Work Distribution (Top 15 Countries)", fontsize=14)
plt.ylabel("Country")
plt.xlabel("Remote Work Type")
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
import pandas as pd
import scipy.stats as stats

# Clean YearsCodePro
def clean_years(x):
    if pd.isna(x):
        return None
    if str(x).lower() == "less than 1 year":
        return 0.5
    if str(x).lower() == "more than 50 years":
        return 51
    try:
        return float(x)
    except:
        return None

df["YearsCodePro_num"] = df["YearsCodePro"].apply(clean_years)
data=df[['JobSat','YearsCodePro_num']].dropna()

#pearson correlation
pearson_corr, pearson_p = stats.pearsonr(data['YearsCodePro_num'],data['JobSat'])
print(f"Pearson correlation: {pearson_corr:.3f}, p-value:{pearson_p:.3e}")

#Spearman correlation 
spearman_corr,spearman_p=stats.spearmanr(data['YearsCodePro_num'],data['JobSat'])
print(f"Spearman Correlation: {spearman_corr: .3f}, p-value: {spearman_p:.3e}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Extract the first employment type before any semicolon
df['Employment_simple'] = df['Employment'].apply(lambda x: x.split(';')[0] if pd.notnull(x) else x)

# Cross-tabulation
employment_edlevel_ct = pd.crosstab(df['EdLevel'], df['Employment_simple'])
print(employment_edlevel_ct)
# Plot
employment_edlevel_ct.plot(kind='bar', stacked=True, figsize=(10,9))
plt.title('Employment Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Status')
plt.tight_layout()
plt.show()
```
